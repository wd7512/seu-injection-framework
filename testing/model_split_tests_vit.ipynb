```python
import torch

# ViT_B_32 expects 3-channel images sized 224x224
batch_size = 1
dummy_input = torch.randn(batch_size, 3, 224, 224)
```

```python
splitter = ViTBlockSplitter(model, dummy_input)
print("Safe split points:", splitter.get_safe_split_points())

for sp in splitter.get_safe_split_points():
    part1_out, part2 = splitter.split_from_point(sp)
    print(f"Split at {sp} successful, part1_out shape: {part1_out.shape}")
```
